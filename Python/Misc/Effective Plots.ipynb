```python
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
v1 = np.random.binomial(n=1, p=0.5, size=200) #np.random.choice(numpy.arange(0, 2), p=[0.5, 0.5], size=200)
v2 = np.random.normal(size=200)
df = pd.DataFrame(np.array([v1,v2]).transpose(), columns=["V1","V2"])
df.astype({"V1":"int"})
```

```python
ax = sns.boxplot(x = "V1", y = "V2", data = df, hue = "V1")
#ax.set_xticklabels(["Good", "Bad"])
legend_handles, _= ax.get_legend_handles_labels()
ax.legend(legend_handles, ["Good", "Bad"], loc="upper center");
```

```python
from matplotlib import rc
rc('mathtext', default='regular') #Write text using Latex

time = np.arange(10)
temp = np.random.random(10)*30
Swdown = np.random.random(10)*100-10
Rn = np.random.random(10)*100-10

fig = plt.figure()
ax = fig.add_subplot(111)

lns1 = ax.plot(time, Swdown, '-', label = 'Swdown')
lns2 = ax.plot(time, Rn, '-', label = 'Rn')
ax2 = ax.twinx()
lns3 = ax2.plot(time, temp, '-r', label = 'temp')

# added these three lines
lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)

ax.grid()
ax.set_xlabel("Time (h)")
ax.set_ylabel(r"Radiation ($MJ\,m^{-2}\,d^{-1}$)")
ax2.set_ylabel(r"Temperature ($^\circ$C)")
ax2.set_ylim(0, 35)
ax.set_ylim(-20,100)
plt.show()
```
