```python
import numpy as np
import scipy as sp
import math
from matplotlib import pyplot as plt
```

```python
#Split normal density

def split_normal_density(x, mu, sigma1, sigma2):

    C = math.sqrt(2/math.pi)/(sigma1 + sigma2)
    
    f_true = C * np.exp(-1/(2*sigma1**2)*(x - mu)**2)
    
    f_false = C * np.exp(-1/(2*sigma2**2)*(x - mu)**2)
    
    mark = np.greater(x, np.ones(x.shape[0])*mu)

    f_true[mark] = f_false[mark]
    
    f = f_true

    return f
```

```python
params_list = [[0, 1, 1], [0, 1, 0.5], [0, 1, 1.5]]

x = np.linspace(-4, 4, 100)

for params in params_list:
    mu, sigma1, sigma2 = params
    y = split_normal_density(x, mu, sigma1, sigma2)
    plt.plot(x, y, '--')
```
